# Cleaning Data

Dropping columns with high missing data, and Interpolating the missing values for remaining columns using some form of regression.

```python
# Homogenizing None
train_df.replace('?', np.nan, inplace=True)
train_df.replace('None', np.nan, inplace=True)
train_df.head()
```

```python
column_stats = train_df.isna().sum()
column_stats = (column_stats / len(train_df)) * 100
column_stats = column_stats.sort_values(ascending=False)
column_stats[column_stats > 0]
```

```python
dropped = []

for col in train_df:
    if train_df[col].isna().sum() > (len(train_df) * .2):
        dropped.append(col)

train_df = train_df.drop(columns=dropped)
train_df.head()
```
